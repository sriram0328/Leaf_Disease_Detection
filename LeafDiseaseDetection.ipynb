```python
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("yolov8s.pt")

# Train the model on your dataset
model.train(data="/content/drive/MyDrive/Plant-leaf-disease-detection.v1i.yolov8/data.yaml", epochs=50, imgsz=640)
```

```python
from ultralytics import YOLO
import cv2
import numpy as np
import PIL.Image
import IPython.display
from google.colab import files
from google.colab.patches import cv2_imshow

# Load trained YOLOv8 model
model = YOLO("/content/runs/detect/train3/weights/best.pt")  # Path to trained model

def upload_and_detect():
    """Upload an image and run leaf disease detection"""
    uploaded = files.upload()  # Upload image file
    for filename in uploaded.keys():
        image_path = filename

        # Load image
        image = cv2.imread(image_path)
        results = model(image)  # Run detection

        # Show results
        # Access the first element of the list, which is the Results object
        results[0].show()
        results[0].save("output.jpg")
        cv2_imshow(cv2.imread("output.jpg"))  # Display in Colab

def capture_and_detect():
    """Capture an image from webcam and run leaf disease detection"""
    from IPython.display import display
    from google.colab.output import eval_js
    from base64 import b64decode
    import cv2
    import numpy as np

    def take_photo(filename='photo.jpg', quality=0.8):
        js = """
        async function takePhoto(quality) {
            const div = document.createElement('div');
            const capture = document.createElement('button');
            capture.textContent = '📸 Capture';
            div.appendChild(capture);
            document.body.appendChild(div);

            const video = document.createElement('video');
            div.appendChild(video);

            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            video.srcObject = stream;
            await video.play();

            google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

            return new Promise((resolve) => {
                capture.onclick = () => {
                    const canvas = document.createElement('canvas');
                    canvas.width = video.videoWidth;
                    canvas.height = video.videoHeight;
                    canvas.getContext('2d').drawImage(video, 0, 0);
                    stream.getTracks().forEach(track => track.stop());
                    div.remove();
                    resolve(canvas.toDataURL('image/jpeg', quality));
                };
            });
        }
        """
        display(IPython.display.Javascript(js))
        data = eval_js('takePhoto({})'.format(quality))
        binary = b64decode(data.split(',')[1])
        with open(filename, 'wb') as f:
            f.write(binary)
        return filename

    # Capture photo
    image_path = take_photo()
    image = cv2.imread(image_path)

    # Perform detection
    results = model(image)
    # Access the first element of the list, which is the Results object
    results[0].show()
    results[0].save("output.jpg")
    cv2_imshow(cv2.imread("output.jpg"))  # Display in Colab

# Choose how to provide the image
print("Choose an option:")
print("1 - Upload an image")
print("2 - Capture from webcam")
option = input("Enter 1 or 2: ")

if option == "1":
    upload_and_detect()
elif option == "2":
    capture_and_detect()
else:
    print("Invalid option! Please enter 1 or 2.")
```

```python
import gradio as gr
from ultralytics import YOLO
import cv2
```

```python
def detect_disease(image):
    """Run leaf disease detection and return the output image."""
    results = model(image)  # Run detection
    results[0].save("output.jpg")  # Save the output image
    return "output.jpg"  # Return the path to the output image
```

```python
# Load the YOLOv8 model (if not already loaded)
model = YOLO("/content/runs/detect/train3/weights/best.pt")

# Create the Gradio interface
iface = gr.Interface(
    fn=detect_disease,
    inputs=gr.Image(type="filepath"),  # Input: image filepath
    outputs=gr.Image(type="filepath"),  # Output: image filepath
    title="Leaf Disease Detection",
    description="Upload an image of a leaf to detect diseases.",
)

# Launch the interface
iface.launch()
```
